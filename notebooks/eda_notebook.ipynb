```python
import pandas as pd
import numpy as np
import os
os.chdir('/Users/pipegalera/dev/ml_diabetes/')
from pyfiles.utils import get_raw_nhanes, clean_nhanes
```

```python
# Extracting data
YEARS = ['2011-2012', '2013-2014']

df = pd.DataFrame()
for year in YEARS:
  df_year = get_raw_nhanes(year).pipe(clean_nhanes)
  df = pd.concat([df, df_year]).reset_index(drop=True)
```

```python
df
```

```python
# Diabetes Definition (Target)
cond_1 = df['Glucose'] >= 7.0
cond_2 = df['Glycohemoglobin'] >= 6.5
cond_3 = df['Diagnosed Diabetes'] == True

df['Diabetes'] = np.where(cond_1 | cond_2 | cond_3, "Yes", "No")
```
